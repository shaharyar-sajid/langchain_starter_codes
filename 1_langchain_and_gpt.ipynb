# Langchain and OpenAI

This is a very basic example on how to use LangChain with an OpenAI chat model to generate text.
It loads your API key from a .env file, initializes the ChatOpenAI model (gpt-5-nano in this case), sends a simple prompt, and prints the modelâ€™s response.

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
```

```python
# create a .env file and set the name of your OpenAI API key like this:
# OPENAI_API_KEY="your_openai_api_key_here"
# The langchain_openai package will automatically read this environment variable.
load_dotenv()
```

```python
llm = ChatOpenAI(model_name="gpt-5-nano")
```

```python
result = llm.invoke("Tell me a joke about programming.")
```

```python
print(result.content)
```
